# Обучение моделей

```python
import pandas as pd
from replay.session_handler import get_spark_session
from replay.data import Dataset
from replay.metrics import Precision, Recall, MAP, MRR, NDCG
from replay.models import ItemKNN
from replay.utils import convert2spark, split_data

# 1. Создаём Spark-сессию (обязательно!)
spark = get_spark_session(cores=4, memory="4g")

# 2. Загружаем данные
interactions = pd.read_csv("interaction_data.csv")
interactions["timestamp"] = pd.to_datetime(interactions["timestamp"])

# 3. Преобразуем в Spark DataFrame
interactions_sdf = convert2spark(interactions)

# 4. Разделяем по времени
train_sdf, test_sdf = split_data(
    interactions_sdf,
    test_size=0.2,
    strategy="time",
    time_column="timestamp"
)

# 5. Обучаем модель
model = ItemKNN(num_neighbours=20, weighting="cosine")
model.fit(train_sdf)

# 6. Получаем рекомендации
recs = model.predict(
    k=10,
    users=test_sdf.select("user_id").distinct(),
    log=train_sdf
)

# 7. Метрики
threshold = 3
metrics = [
    Precision(k=10, threshold=threshold),
    Recall(k=10, threshold=threshold),
    MAP(k=10),
    MRR(k=10),
    NDCG(k=10)
]

results = {}
for metric in metrics:
    results[metric.__class__.__name__] = metric(test_sdf, recs)

print(results)
```
